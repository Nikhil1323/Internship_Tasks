# Outliers

## Outliers are those values in the dataset that are at very high deviation from the mean value and can cause significant disruption in the analysis of data 

```python
# The outliers in our dataset can be present in columns whose datatypes are int64 or float64
```

```python
df2.dtypes
```

```python
#to check outliers in the age column we use the Z score method and the formula Z = (i-mean)/standard deviation.
age = []
for i in df2['Age']:
    age.append(i)
```

# We use Z Score method to find the outliers

```python
outliers =[]
def detect_outliers(data):
    
    threshold = 3
    mean = np.mean(data)
    std = np.std(data)
    
    for i in data:
        z_score = (i - mean)/ std
        if np.abs(z_score) > threshold:
            outliers.append(i)
    return outliers
```

```python
outlier_pt = detect_outliers(age)
```

```python
outlier_pt
```

## Using the above method we did not find any outlier in the Age column.

```python
#to check outliers in the Fare column we use the Z score method and the formula Z = (i-mean)/standard deviation.
ticket_price = []
for i in df2['Fare']:
    ticket_price.append(i)
```

```python
outlier_fare = detect_outliers(ticket_price)
```
